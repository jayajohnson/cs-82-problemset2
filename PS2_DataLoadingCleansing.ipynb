```
import bs4
import os
import codecs

directory = "/content/factiva";
outdir = "/content/factivatxt/"
print("Validate files were loaded!")
list = os.listdir(directory) # dir is your directory path
number_files = len(list)
print("Number of files extracted should be 8863!")
print(number_files)

for filename in os.listdir(directory):
    htmlfile = directory+"/"+filename
    f=codecs.open(htmlfile, 'r', 'utf-8')
    soup = bs4.BeautifulSoup(f)
    data = soup.get_text('\n')
    ##print(data)
    outputfilename = outdir+filename+".txt"
    ##print(outputfilename)
    wf = open(outputfilename+".txt","w+");
    wf.write(data);
    wf.close();
    
print("Validate files were stripped and converted to txt!")
list = os.listdir(outdir) # dir is your directory path
number_files = len(list)
print("Number of files extracted should be 8863!")
print(number_files)
```
